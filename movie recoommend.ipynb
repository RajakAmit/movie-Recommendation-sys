```python
import ast

# Corrected convert function
def convert(obj):
    l = []
    try:
        for i in ast.literal_eval(obj):  # Convert the string to a Python object
            l.append(i['name'])         # Append the 'name' field to the list
        return l
    except (ValueError, SyntaxError, TypeError):  # Handle invalid or None values
        return None  # Return None if the input is not a valid string

# Apply the function to the 'genres' column
movie['genres'] = movie['genres'].apply(convert)

# Display the transformed column
print(movie['genres'])
```

```python
movie.head()
```

```python
movie['keywords']=movie['keywords'].apply(convert)
```

```python
#find top 3 actors from dictonaries to finf actual name 

def convert3(obj):
    l=[]
    counter =0
    for i in ast.literal_eval(obj):
        if counter != 3:
            l.append(i['name'])
            counter+=1
        else:
            break
    return l
```

```python
movie['cast']=movie['cast'].apply(convert3)
```

```python
movie.head()
```

```python
def director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l
```

```python
movie['crew']=movie['crew'].apply(director)
```

```python


# Apply the transformation
movie['genres'] = movie['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
movie['keywords'] = movie['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
movie['cast'] = movie['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
movie['crew'] = movie['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
#convert list of tags into string 
df['tags']=df['tags'].apply(lambda x:" ".join(x))
```

```python
df['tags']=df['tags'].apply(lambda x:x.lower())
```

```python
df.head()
```

```python
# stemming and removing stop words 
import nltk 
```

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
from nltk.stem import PorterStemmer
```

```python
stop_words=set(stopwords.words('english'))
```

```python
stemmer=PorterStemmer()
```

```python
# Apply stemming to words in 'tags'
df['tags'] = df['tags'].apply(lambda x: ' '.join([stemmer.stem(word) for word in x.split() if word not in stop_words]))
```

```python
df['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Initialize CountVectorizer with the desired parameters
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vector=cv.fit_transform(df['tags']).toarray()
```

```python
vector
```

```python
vector.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Get feature names
feature_names = cv.get_feature_names_out()
print(feature_names)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
def recommend(movie):
    movie_index = df[df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movie_list= sorted(list(enumerate(distances)),reverse=True,key = lambda x: x[1])[1:6]
    
    for i in movie_list:
        print(i[0])
```
